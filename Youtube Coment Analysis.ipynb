## 2. Preprocessing: 

```python
# Remove Emojis
def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F700-\U0001F77F"  # alchemical symbols
        u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        u"\U0001FA00-\U0001FA6F"  # Chess Symbols
        u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        u"\U00002702-\U000027B0"  # Dingbats
        u"\U000024C2-\U0001F251" 
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['comment_text'] = df['comment_text'].apply(remove_emojis)
```

```python
# Standardize Text Data
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[\d\[\]]', '', text)  # Remove numbers and brackets
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()  # Remove leading and trailing spaces
    return text

df_cleaned['comment_text'] = df_cleaned['comment_text'].apply(clean_text)
```

```python
def clean_comment_text(text):
    # Remove newline characters
    text = text.replace('\n', ' ')
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Convert to lowercase
    text = text.lower()
    return text
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Function to classify sentiment based on VADER's compound score
def classify_sentiment(comment):
    scores = sid.polarity_scores(comment)
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# nltk.download('stopwords')
# nltk.download('wordnet')
```

```python
# Preprocess text data
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    tokens = re.findall(r'\b\w+\b', text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    return ' '.join(tokens)

cleaned_df['processed_text'] = cleaned_df['comment_text'].apply(preprocess_text)
```

```python
cleaned_df.tail()
```

```python
# Clean the comment text
def preprocess_text(text):
    tokens = re.findall(r'\b\w+\b', text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    return ' '.join(tokens)
cleaned_df['comment_text'] = cleaned_df['comment_text'].apply(clean_comment_text)
```

```python
# Convert text to document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(cleaned_df['processed_text'])
```

```python
# Perform LDA
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(dtm)
```

```python
# Display the topics
def display_topics(model, feature_names, no_top_words):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        topics.append([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]])
        print(" ".join(topics[-1]))
    return topics
```

```python
no_top_words = 10
tf_feature_names = vectorizer.get_feature_names_out()
topics = display_topics(lda, tf_feature_names, no_top_words)
```

```python

# Save the topics into a DataFrame
topic_dict = {'Topic_' + str(i+1): topics[i] for i in range(n_topics)}
topics_df = pd.DataFrame.from_dict(topic_dict, orient='index', columns=[f'Word_{i+1}' for i in range(no_top_words)])
print(topics_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Correlate Sentiment and Engagement
# Aggregate engagement metrics by sentiment
sentiment_engagement = cleaned_df.groupby('sentiment')[['likes', 'replies']].mean().reset_index()
print(sentiment_engagement)

# Visualize the relationship
plt.figure(figsize=(12, 6))
sns.barplot(data=sentiment_engagement, x='sentiment', y='likes')
plt.title('Average Likes by Sentiment')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=sentiment_engagement, x='sentiment', y='replies')
plt.title('Average Replies by Sentiment')
plt.show()
```

```python
# Assign topics to comments
topic_assignments = lda.transform(dtm)
cleaned_df['topic'] = topic_assignments.argmax(axis=1)
```

```python
# Correlate Topics and Engagement

# Aggregate engagement metrics by topic
topic_engagement = cleaned_df.groupby('topic')[['likes', 'replies']].mean().reset_index()
print(topic_engagement)

```

```python
# Visualize the relationship
plt.figure(figsize=(12, 6))
sns.barplot(data=topic_engagement, x='topic', y='likes')
plt.title('Average Likes by Topic')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=topic_engagement, x='topic', y='replies')
plt.title('Average Replies by Topic')
plt.show()
```
